```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('fin_data_of_proj.xlsx')
df = df[:-11]
df
```

```python
df = df.drop(['Unnamed: 0', 'application_name', 'author_id', 'contractor_id', 'designer_id', 'engine_id', 'status',
              'decision_date', 'region_name', 'type', 'turn', 'period_month', 'build_qty', 'share_cost', 'comm_cost',
              'prolongation', 'state_id', 'annul_date', 'state_name', 'дата фин данных', 'percent_of_lifi_cycle'], axis=1)
```

## Models Testing

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, 
    roc_auc_score, roc_curve
)

# Define the target variable (y) and features (X)
y = df['Target']
X = df.drop(columns=['Target'])
X.replace([np.inf, -np.inf], np.nan, inplace=True)

# Identify non-numeric columns
non_numeric_cols = X.select_dtypes(include=['object', 'datetime', 'timedelta']).columns

# Apply LabelEncoder to non-numeric columns           
label_encoder = LabelEncoder()
for col in non_numeric_cols:
    X[col] = label_encoder.fit_transform(X[col].astype(str))

# Handle missing values using SimpleImputer (fill with median)
imputer = SimpleImputer(strategy='median')
X_imputed_encoded = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed_encoded, y, test_size=0.2, random_state=42)

# Initialize ensemble models
models = {
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Bagging": BaggingClassifier(),
}

# Train models and collect predictions and feature importances
feature_importances = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    
    # Store feature importance if available
    if hasattr(model, 'feature_importances_'):
        feature_importances[name] = model.feature_importances_

    # Predict and evaluate
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    print(f"{name} Accuracy: {accuracy_score(y_test, y_pred):.2f}")
    print(f"{name} Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}\n")
    print(f"{name} ROC AUC: {roc_auc_score(y_test, y_prob):.2f}\n")


plt.tight_layout()
plt.show()
```

## Feature Importance


```python
from matplotlib.backends.backend_pdf import PdfPages

num_features = X.shape[1]
fig, axes = plt.subplots(len(feature_importances), 1, figsize=(40, 20 * len(feature_importances)))

with PdfPages("feature_importances.pdf") as pdf:
    for idx, (name, importance) in enumerate(feature_importances.items()):
        axes[idx].barh(X.columns, importance)
        axes[idx].set_title(f"{name} Feature Importance")
        axes[idx].set_xlabel("Importance Score")
        axes[idx].set_ylabel("Features")
        
    pdf.savefig(fig)
    plt.close(fig) 
```
