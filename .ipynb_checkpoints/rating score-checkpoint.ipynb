```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Full_Data_With_Life_Bin_Backlog.xlsx')
df
```

```python
df.info()
```

```python
df = df[df['Backlog'] == '30%+']
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score


# Define the target variable (y) and features (X)
y = df['Life_Bin']
X = df.drop(columns=['Life_Bin'])

# Identify non-numeric columns
non_numeric_cols = X.select_dtypes(include=['object', 'datetime', 'timedelta']).columns

# Apply LabelEncoder to non-numeric columns
label_encoder = LabelEncoder()
for col in non_numeric_cols:
    X[col] = label_encoder.fit_transform(X[col].astype(str))

# Handle missing values using SimpleImputer (fill with median)
imputer = SimpleImputer(strategy='median')
X_imputed_encoded = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed_encoded, y, test_size=0.4, random_state=42)

# Initialize ensemble models
rf_model = RandomForestClassifier()
gb_model = GradientBoostingClassifier()
ada_model = AdaBoostClassifier()

# Train the models
rf_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
ada_model.fit(X_train, y_train)

# Get predictions
rf_pred = rf_model.predict(X_test)
gb_pred = gb_model.predict(X_test)
ada_pred = ada_model.predict(X_test)

# Print classification reports
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
print("Gradient Boosting Accuracy:", accuracy_score(y_test, gb_pred))
print("AdaBoost Accuracy:", accuracy_score(y_test, ada_pred))
```
